```python
patient_folder = "BRATS/ASNR-MICCAI-BraTS2023-GLI-Challenge-TrainingData/BraTS-GLI-00000-000"

modalities = ['t1c', 't1n', 't2f', 't2w', 'seg']
images = {}
```

```python
for mod in modalities:
    img_path = os.path.join(patient_folder,
                            f"{os.path.basename(patient_folder)}-{mod}.nii.gz")
    img = nib.load(img_path)
    images[mod] = img.get_fdata()

seg_path = os.path.join(patient_folder, f"{os.path.basename(patient_folder)}-seg.nii.gz")
seg_img = nib.load(seg_path)
seg = seg_img.get_fdata()
```

```python
# Choose a slice index to visualize
slice_idx = images['t2f'].shape[2] // 2  # middle slice
fig, axes = plt.subplots(1, 5, figsize=(20, 5))

for i, mod in enumerate(modalities):
    axes[i].imshow(images[mod][:, :, slice_idx], cmap='gray')
    axes[i].set_title(mod.upper())
    axes[i].axis('off')

# Plot segmentation mask
axes[4].imshow(seg[:, :, slice_idx], cmap='jet', alpha=0.5)
axes[4].set_title("Segmentation")
axes[4].axis('off')

plt.show()
```

```python
# Replace this with the path to your extracted dataset
data_path = "BRATS/ASNR-MICCAI-BraTS2023-GLI-Challenge-TrainingData/"
sample_case = os.listdir(data_path)[0]
sample_path = os.path.join(data_path, sample_case)

# Load all modalities
modalities = {
    "T1n": nib.load(os.path.join(sample_path, f"{sample_case}-t1n.nii.gz")).get_fdata(),
    "T1c": nib.load(os.path.join(sample_path, f"{sample_case}-t1c.nii.gz")).get_fdata(),
    "T2w": nib.load(os.path.join(sample_path, f"{sample_case}-t2w.nii.gz")).get_fdata(),
    "T2f": nib.load(os.path.join(sample_path, f"{sample_case}-t2f.nii.gz")).get_fdata(),
    "Seg": nib.load(os.path.join(sample_path, f"{sample_case}-seg.nii.gz")).get_fdata().astype(int)
}
```

```python
for name, image in modalities.items():
    print(f"{name} shape: {image.shape}, dtype: {image.dtype}")
```

```python
def show_mid_slice(img, title="", cmap="gray"):
    mid = img.shape[2] // 2
    plt.imshow(img[:, :, mid], cmap=cmap)
    plt.title(title)
    plt.axis('off')

plt.figure(figsize=(15, 6))
for i, (name, img) in enumerate(modalities.items()):
    plt.subplot(1, 5, i + 1)
    show_mid_slice(img, title=name, cmap="gray" if name != "Seg" else "nipy_spectral")
plt.tight_layout()
plt.show()
```

```python
# In a new script or notebook, after running preprocess.py

import numpy as np
import matplotlib.pyplot as plt
import os

# Path to the folder where you saved the .npy files
PROCESSED_DATA_PATH = "BRATS-data/processed/"

# Get a list of the processed files
all_image_files = os.listdir(PROCESSED_DATA_PATH)
example_img_file = all_image_files[0] # e.g., 'BraTS-GLI-00518-001_img.npy'
example_seg_file = example_img_file.replace('_img.npy', '_seg.npy')

# --- Load the .npy files ---
image_array = np.load(os.path.join(PROCESSED_DATA_PATH, example_img_file))
seg_array = np.load(os.path.join(PROCESSED_DATA_PATH, example_seg_file))

# --- Print information to understand the data ---
print(f"--- Image Array ---")
print(f"File: {example_img_file}")
print(f"Shape: {image_array.shape}")      # Expected output: (4, 240, 240, 155)
print(f"Data Type: {image_array.dtype}")  # Expected output: float32

print(f"\n--- Segmentation Array ---")
print(f"File: {example_seg_file}")
print(f"Shape: {seg_array.shape}")        # Expected output: (240, 240, 155)
print(f"Data Type: {seg_array.dtype}")    # Expected output: uint8
print(f"Unique labels: {np.unique(seg_array)}") # See the label values (e.g., [0, 1, 2, 3])
```

```python
# --- Visualize a slice to confirm it looks right ---
slice_index = 80 # Pick a slice from the middle

# The image_array has 4 channels. Let's visualize the 2nd channel (T1c).
t1c_slice = image_array[1, :, :, slice_index]

# The segmentation array has no channel dimension.
seg_slice = seg_array[:, :, slice_index]

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(t1c_slice, cmap='gray')
axes[0].set_title(f'T1c Scan (Slice {slice_index})')
axes[0].axis('off')

axes[1].imshow(seg_slice, cmap='jet') # 'jet' colormap shows different labels in different colors
axes[1].set_title(f'Segmentation Mask (Slice {slice_index})')
axes[1].axis('off')

plt.show()
```
